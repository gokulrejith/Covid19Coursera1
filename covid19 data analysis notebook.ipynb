### Task 2.2: Delete the useless columns

```python
dataset.drop(["Lat", "Long"], axis = 1, inplace = True)
```

### Task 2.3: Aggregating the rows by the country

```python
datasetAggregated = dataset.groupby("Country/Region").sum()
```

### task 3.2: find maxmimum infection rate for China

### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries = list(datasetAggregated.index)
MaxInfectionRates = []
for c in countries:
    MaxInfectionRates.append(datasetAggregated.loc[c].diff().max())
datasetAggregated["MaxInfectionRate"] = MaxInfectionRates
```

```python
datasetAggregated
```

### Task 3.4: create a new dataframe with only needed column 

```python
cdata = pd.DataFrame(datasetAggregated["MaxInfectionRate"])
```

```python
cdata
```
